```python
#Import depedencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

app_ship = pd.read_csv("ApplianceShipments.csv")
app_ship.head()
app_ship_plot = pd.Series(app_ship.Shipments.values, index = app_ship.Quarter)
app_ship_plot.plot(legend=True)
plt.xlabel("Quarters")
plt.ylabel("Shipments in Million of USD")
#B) Zoom in 3500-5000
plt.ylim(3500, 5000)
#plt.plot(app_ship.quarter, app_ship.shipments)
#A) Show graph
plt.show()
```

```python
#C)
quarter_1 = app_ship[app_ship["Quarter"].str.contains("Q2|Q3|Q4",) == False]
quarter_1 = pd.Series(quarter_1.Shipments.values, index = quarter_1.Quarter)

quarter_2 = app_ship[app_ship["Quarter"].str.contains("Q1|Q3|Q4",) == False]
quarter_2 = pd.Series(quarter_2.Shipments.values, index = quarter_2.Quarter)

quarter_3 = app_ship[app_ship["Quarter"].str.contains("Q2|Q1|Q4",) == False]
quarter_3 = pd.Series(quarter_3.Shipments.values, index = quarter_3.Quarter)

quarter_4 = app_ship[app_ship["Quarter"].str.contains("Q2|Q3|Q1",) == False]
quarter_4 = pd.Series(quarter_4.Shipments.values, index = quarter_4.Quarter)


quarter_1.plot(label="Quarter 1", legend = True)
quarter_2.plot(label="Quarter 2", legend = True)
quarter_3.plot(label="Quarter 3", legend = True)
quarter_4.plot(label="Quarter 4", legend = True)
plt.ylim(3500, 5000)
plt.show()
#df = df[df["team"].str.contains("Team 1") == False]
#df = df.drop(df[df['product_name'] == 'Coca Cola'].index)

```

#### C) It seems as if Quarter 1 and 4 has the lowest amount of shipments and Quarter 2 has the highest

```python
#D)
#print(app_ship["Quarter"].str.contains("Q2|Q3|Q1",) == False)
app_ship_plot.head()
app_ship_yearly = pd.Series(app_ship.Shipments.values, index=app_ship["Quarter"].str[3:])
#app_ship_yearly.head()
app_ship_yearly_plot = app_ship_yearly.groupby(level=0).sum()

app_ship_yearly_plot.plot(xlabel= "Year")
plt.show()
```

## Oppgave 3.2

```python
df = pd.read_csv("RidingMowers.csv")

df.head()

#Scatter plot with colour coded Ownership

fig, ax = plt.subplots()

for pl, colour in ("Nonowner", "C1"), ("Owner", "C0"):
    sub_df = df[df.Ownership == pl]
    ax.scatter(sub_df.Income, sub_df.Lot_Size, color="none", edgecolor=colour)
ax.set_xlabel("Income")
ax.set_ylabel("Lot Size")
ax.legend(["Owner", "Non-Owner"])
plt.show()
```

```python
#a)
laptop_jan_df = pd.read_csv("LaptopSalesJanuary2008.csv")

laptop_jan_df = pd.DataFrame({
    "Store" : laptop_jan_df.pop("Store Postcode"),
    "Retail" : laptop_jan_df.pop("Retail Price")
})

laptop_jan_df.head()

#Groupby and find mean
group_laptop = laptop_jan_df.groupby("Store")
group_laptop.mean()
group_laptop_mean = group_laptop.mean()

ax = group_laptop_mean.plot.bar(ylim=[480,500])
plt.show()

print(group_laptop_mean.idxmax(),group_laptop_mean.max())
#Highest average is 495 at store N17 6QA

print(group_laptop_mean.idxmin(), group_laptop_mean.min())
#Lowest average is 481 at store W4 3PH
```

```python
#b)

laptop_jan_df.boxplot(column="Retail", by="Store",vert=False)

plt.show()
```

The N7 seems to be more expensive overall while W4 is cheaper, but both seems to be distributed normally
